```python
main_path = '.\data\ptsd'
```

```python
test_df = pd.read_csv('./testset_CSV.csv')
test_df.head(20)
```

```python
classes_reverse = {
            'Compulsory Keep BothSide':0,
            'Compulsory Keep Left':1, 
            'Compulsory Keep Right':2, 
            'Cycle crossing':3, 
            'Danger':4, 
            'Give Way':5, 
            'Hump':6, 
            'Left Bend':7, 
            'Left Margin':8, 
            'Left Turn Prohibited':9, 
            'Maximum Speed 30':10, 
            'Maximum Speed 40':11, 
            'Maximum Speed 50':12, 
            'Maximum Speed 60':13, 
            'Maximum Speed 70':14, 
            'Maximum Speed 80':15, 
            'Maximum Speed 90':16, 
            'MotorCycle Prohibited':17, 
            'No Entry':18, 
            'No Horn':19, 
            'NO Stopping':20, 
            'NO Waiting':21, 
            'One way Traffic':22, 
            'Park':23, 
            'Park Forbidden':24, 
            'Pedestrain':25, 
            'Pedestrian crossing':26, 
            'Right Bend':27, 
            'Right Margin':28, 
            'Right Turn Prohibited':29, 
            'Road Work':30,
            'Roundabouts':31, 
            'School':32, 
            'School Crossing':33, 
            'Side Road Right':34, 
            'Slow':35, 
            'Speed Camera':36, 
            'STOP':37, 
            'Truck Prohibited':38, 
            'Two Way Traffic':39, 
            'U-Turn':40, 
            'U-Turn Allowed':41, 
            'U-turn Prohibited':42
}
```

```python
os.mkdir(os.path.join(main_path,"PTSD_Recognition","sep_test"))

for index, row in test_df.iterrows():
    class_idx = classes_reverse[row['species']]
    image_name = "PTSD_" + row['file']
    
    image_dir = os.path.join(main_path,"PTSD_Recognition","test",image_name)
    class_dir = os.path.join(main_path,"PTSD_Recognition","sep_test","{:02d}".format(class_idx))
             
    if os.path.exists(class_dir):
        shutil.copyfile(image_dir, os.path.join(class_dir,image_name))
    else:
        os.mkdir(class_dir)
        shutil.copyfile(image_dir, os.path.join(class_dir,image_name))
```
